```python
# Ingreso total por cada tienda
'''
En este primer análisis, debes calcular el ingreso total de cada tienda.
Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.
'''

ingreso_total1 = tienda1["Precio"].sum()
ingreso_total2 = tienda2["Precio"].sum()
ingreso_total3 = tienda3["Precio"].sum()
ingreso_total4 = tienda4["Precio"].sum()

print(f"El ingreso total de la tienda 1 es de: {ingreso_total1}")
print(f"El ingreso total de la tienda 2 es de: {ingreso_total2}")
print(f"El ingreso total de la tienda 3 es de: {ingreso_total3}")
print(f"El ingreso total de la tienda 4 es de: {ingreso_total4}")
```

# 2. Ventas por categoría

```python
# Ventas por categoría

'''
En este debes calcular la cantidad de productos vendidos por categoría en cada tienda.
La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo,
mostrando las categorías más populares de cada tienda.
'''

ventas_por_categoria1 = tienda1["Categoría del Producto"].value_counts().reset_index()
ventas_por_categoria1.columns = ['Categoría', 'Cantidad']

ventas_por_categoria2 = tienda2["Categoría del Producto"].value_counts().reset_index()
ventas_por_categoria2.columns = ['Categoría', 'Cantidad']

ventas_por_categoria3 = tienda3["Categoría del Producto"].value_counts().reset_index()
ventas_por_categoria3.columns = ['Categoría', 'Cantidad']

ventas_por_categoria4 = tienda4["Categoría del Producto"].value_counts().reset_index()
ventas_por_categoria4.columns = ['Categoría', 'Cantidad']


print("Ventas Tienda 1:\n", ventas_por_categoria1, "\n")
print("Ventas Tienda 2:\n", ventas_por_categoria2, "\n")
print("Ventas Tienda 3:\n", ventas_por_categoria3, "\n")
print("Ventas Tienda 4:\n", ventas_por_categoria4, "\n")
```

# 3. Calificación promedio de la tienda


```python
# Valoración media por tienda
'''
En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda.
El objetivo es conocer la satisfacción del cliente con los productos vendidos.
'''

calificacion_media1 = tienda1["Calificación"].mean()
calificacion_media2 = tienda2["Calificación"].mean()
calificacion_media3 = tienda3["Calificación"].mean()
calificacion_media4 = tienda4["Calificación"].mean()

print(f"La calificación media de la tienda 1 es de: {calificacion_media1:.2f}")
print(f"La calificación media de la tienda 2 es de: {calificacion_media2:.2f}")
print(f"La calificación media de la tienda 3 es de: {calificacion_media3:.2f}")
print(f"La calificación media de la tienda 4 es de: {calificacion_media4:.2f}")
```

```python
# Gráfico de líneas - Calificación promedio por tienda

calificaciones = [calificacion_media1, calificacion_media2, calificacion_media3, calificacion_media4]

plt.figure(figsize=(8, 6))
plt.plot(tiendas, calificaciones, marker='o', linestyle='-', color='orange')
plt.title("Calificación Promedio por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Calificación Promedio")
plt.ylim(0, 5)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
# Valor del envío promedio por tienda
'''
En este paso, debes calcular el costo de envío promedio para cada tienda.
El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.
'''

envio_promedio1 = tienda1["Costo de envío"].mean()
envio_promedio2 = tienda2["Costo de envío"].mean()
envio_promedio3 = tienda3["Costo de envío"].mean()
envio_promedio4 = tienda4["Costo de envío"].mean()

print(f"El envío promedio de la tienda 1 es de: {envio_promedio1:.2f}")
print(f"El envío promedio de la tienda 2 es de: {envio_promedio2:.2f}")
print(f"El envío promedio de la tienda 3 es de: {envio_promedio3:.2f}")
print(f"El envío promedio de la tienda 4 es de: {envio_promedio4:.2f}")
```

```python
# Gráfico de torta: Valor del envío promedio por tienda

tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
envios = [envio_promedio1, envio_promedio2, envio_promedio3, envio_promedio4]

plt.figure(figsize=(7, 7))
plt.pie(envios, labels=tiendas, autopct='%1.1f%%', startangle=90, colors=plt.cm.Set3.colors)
plt.title("Proporción del Costo Promedio de Envío por Tienda")
plt.axis('equal')  # Hace que el gráfico sea circular
plt.tight_layout()
plt.show()
```
